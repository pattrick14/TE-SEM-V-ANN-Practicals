```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([[1, 1],[1, -1],[-1, 1],[-1, -1]])
y = np.array([-1, -1, -1, 1])
```

```python
w = np.zeros(x.shape[1]) #Returns shape of array
b = 0
alpha = 0.3
```

```python
for _ in range(20):
    for i in range(x.shape[0]):
        y_pred = np.sign(np.dot(x[i], w)+b)

        if y_pred != y[i]:
            w += alpha * y[i] * x[i]
            b += alpha * y[i]

x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
plt.scatter(x[:, 0], x[:, 1])

slope = -w[0] / w[1]
intercept = -b / w[1]
plt.plot([x_min, x_max], [x_min * slope + intercept, x_max * slope + intercept])

plt.show()
```
